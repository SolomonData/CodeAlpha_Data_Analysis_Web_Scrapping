# Average Quote Lenght by Authors

```python
avg_quote_len = df.groupby("Author")["Quote_Length"].mean().reset_index()
avg_quote_len.columns = ["Author", "Avg_Quote_Length"]
avg_quote_len["Avg_Quote_Length"] = avg_quote_len["Avg_Quote_Length"].round(2)
avg_quote_len = avg_quote_len.sort_values("Avg_Quote_Length", ascending = False)

avg_quote_len
```

# Total Tags and Unique Tags


```python
# How many Tags are there?

total_tags = []

for tag_string in df["Tags"]:
    if pd.notna(tag_string):    
        tag_list = tag_string.split(", ")
        total_tags.extend(tag_list)

print(f"Total Number of Tags: {len(total_tags)}")
```

```python
# How many Unique Tags are there?
unique_tags = sorted(set(total_tags))
print(f"Total Number of Unique Tags: {len(unique_tags)}")
```

```python
for i, tag in enumerate(unique_tags, start=1):
    print(f"{i}. {tag}")
```

## How many times does each tag appear?
## Which tag has the most appearance?

```python
for i, tag in enumerate(sorted(unique_tags), start = 1):
    print(f"{i}. {tag}")
```

```python

from collections import Counter

tag_counts = Counter(total_tags)

tag_counts_df = (
    pd.DataFrame(tag_counts.items(), columns=["Tag", "Tags_Frequency"])
    .sort_values(by="Tags_Frequency", ascending=False)
    .reset_index(drop=True)
)

tag_counts_df.insert(0, "S/N", range(1, len(tag_counts_df) + 1))

print(tag_counts_df)
```

# 1. Are most quotes relatively short?

# Null Hypothesis (H₀)
The majority of quotes contain more than 25 words.
    
# Alternative Hypothesis (H₁)
The majority of quotes contain 25 or fewer words.

```python
df.describe().round(2)
```

```python
import matplotlib.pyplot as plt



plt.hist(df["Quote_Length"], bins=10)
plt.xlabel("Number of Words(Quote_Length)")
plt.ylabel("Number of Quotes")
plt.title("Distribution of Quote Length")
plt.show()

print("The histogram shows that most quotes contain fewer than 25 words. Therefore, the data supports the alternative hypothesis.")
```

# 2. Do a small number of authors contribute most of the quotes?

# Null Hypotheses(H₀)
The Quotes are evenly distributed.

# Alternative Hypothesis(H₁)
A few authors has more quotes.

```python
author_quotes_counts
```

# Statistics

```python
author_quotes_counts["Total_Quotes"].describe().round(2)
```

# Visualization

```python
plt.figure(figsize=(16,8))

plt.bar(author_quotes_counts["Author"], author_quotes_counts["Total_Quotes"])

plt.xticks(rotation=45, ha="right")

plt.xlabel("Author")
plt.ylabel("Number of Quotes")
plt.title("Authors by Number of Quotes")

plt.show()

one_quote = (author_quotes_counts["Total_Quotes"] == 1).sum()
percentage = (one_quote / len(author_quotes_counts)) * 100
print(f"Total_Authors with just one quote: {one_quote}")
print(f"Percentage of Authors with just one quote: {percentage}%")

print("The hypothesis is supported. Descriptive statistics show that the median number of quotes per author is one, indicating that at least half of the authors contributed only a single quote. Furthermore, 70% of authors appear only once in the dataset, while a small number of authors contribute several quotes. The bar chart also illustrates that a few authors dominate the dataset, confirming that quote contributions are not evenly distributed.")
```

# 3. Are love tags more common than books tags?

# Null Hypothesis (H₀)
Love tags are not more frequent than books tags.

# Alternative Hypothesis (H₁)
Love tags are more frequent than books tags.

```python
tag_counts_df
```

# Statistics

```python
tag_counts_df["Tags_Frequency"].describe().round(2)
```

# Visualization

```python
top_tags = tag_counts_df.head(10)

plt.figure(figsize=(7,6))

plt.bar(top_tags["Tag"], top_tags["Tags_Frequency"])

plt.xticks(rotation=45, ha="right")

plt.xlabel("Tag")
plt.ylabel("Frequency")
plt.title("Top 10 Tags Frequency")

plt.show()

total_frequency = tag_counts_df["Tags_Frequency"].sum()

love_frequency = tag_counts_df.loc[
    tag_counts_df["Tag"] == "love",
    "Tags_Frequency"
].iloc[0]
books_frequency = tag_counts_df.loc[
    tag_counts_df["Tag"] == "books",
    "Tags_Frequency"
].iloc[0]

love_percentage = (love_frequency / total_frequency)*100
books_percentage = (books_frequency / total_frequency)*100

print(f"Total Frequency: {total_frequency}")
print(f"Love Frequency: {love_frequency} occurence({love_percentage.round(2)}%)")
print(f"Books Frequency: {books_frequency} occurence({books_percentage.round(2)}%)")
print("The descriptive statistics and visualization support the hypothesis that the love tag appears more frequently than the books tag.")
```
