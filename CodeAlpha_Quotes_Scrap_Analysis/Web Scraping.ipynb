```python
quotes = []
authors = []
tags = []

for page_num in range(1, 11):

    url = f"http://quotes.toscrape.com/page/{page_num}/"

    page = requests.get(url)

    soup = BeautifulSoup(page.text, "html.parser")

    quote_blocks = soup.find_all("div", class_="quote")

    for block in quote_blocks:

        quote = block.find("span", class_="text").get_text().strip()

        author = block.find("small", class_="author").get_text().strip()

        tag_list = [
            tag.get_text().strip()
            for tag in block.find_all("a", class_="tag")
        ]

        quotes.append(quote)
        authors.append(author)
        tags.append(", ".join(tag_list))
```

```python
len(quotes)
len(authors)
len(tags)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Quote": quotes,
    "Author": authors,
    "Tags": tags
})
pd.set_option("display.max_rows", 100)
df
```
